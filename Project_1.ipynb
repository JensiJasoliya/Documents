```python
def Del_Char(WORD):
    # Generate a list of words by deleting one character from the input WORD
    return [WORD[:i] + WORD[i+1:] for i in range(len(WORD))]

# Define the input word
Name = 'pace'

# Call the Del_Char function to generate outputs by deleting characters from the input word
OP = Del_Char(WORD=Name)

# Print the original input word
print("Input Word:", Name)

# Print the list of generated words after deleting each character
print("Output:", OP)
```

```python
def switch_char(word):
    # Initialize an empty list to store the words generated by switching characters
    switched_words = []

    # Loop through each character in the word except the last one
    for i in range(len(word) - 1):
        # Create a new word by switching the current character with the next character
        new_word = word[:i] + word[i + 1] + word[i] + word[i + 2:]
        # Append the new word to the list of switched words
        switched_words.append(new_word)

    # Return the list of switched words
    return switched_words

# Define a product name to test the function
Product_Name = 'pace'
# Call the switch_char function with the defined product name
OP = switch_char(Product_Name)

# Print the input word
print("Input Word:", Product_Name)
# Print the output list of words generated by switching characters
print("Output:", OP)
```

```python
import string  # Import the string module to access string constants

def ins_char(word):
    # Create a string containing all lowercase letters
    alphabet = string.ascii_lowercase

    # Initialize an empty list to store the resulting words
    result = []  # NOTE: Fixed the empty return list to actually use it for storing results

    # Loop through each position in the word (from 0 to len(word))
    for i in range(len(word) + 1):
        # Loop through each letter in the alphabet
        for letter in alphabet:
            # Create a new word by inserting the current letter at the current position
            new_word = word[:i] + letter + word[i:]
            # Append the new word to the results list
            result.append(new_word)

    # Return the list of new words created by inserting letters
    return result

# Call the ins_char function with the example word 'at'
OP = ins_char('abc')

# Print the output of the function
print("Output:", OP)

# Print the total number of generated strings
print(f"Total number of strings: {len(OP)}")
```

# Edit Distance Calculations



```python
def edit_dis_one(word, allow_switches=True):
    let = 'abcdefghijklmnopqrstuvwxyz'

#Create a set of edits by deleting each character in the word
    edits = {word[:i] + word[i+1:] for i in range(len(word))}
#Insert every letter at every possible position in the word
    edits |= {word[:i] + c + word[i:] for i in range(len(word) + 1) for c in let}
#Replace each character in the word with every letter of the alphabet
    edits |= {word[:i] + c + word[i+1:] for i in range(len(word)) for c in let}
# If allowed, swap adjacent characters in the word
    if allow_switches:
        edits |= {word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(len(word) - 1)}

    return sorted(edits - {word}) # Exclude the original word and return sorted edits

word = 'jensi'
result = edit_dis_one(word)
print(f"Edit distance one for '{word}':")
print(result)
print(f"Number of results: {len(result)}")
```

```python
def edit_dis_one(word, allow_switches=True):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    n = len(word)
    edits = []

    # Deletions
    edits += [word[:i] + word[i+1:] for i in range(n)]
    # Insertions
    edits += [word[:i] + c + word[i:] for i in range(n+1) for c in letters]
    # Replacements
    edits += [word[:i] + c + word[i+1:] for i in range(n) for c in letters]
    # Transpositions
    if allow_switches:
        edits += [word[:i] + word[i+1] + word[i] + word[i+2:] for i in range(n-1)]

    return set(edits)

def edit_dis_two(word, allow_switches=True):
    edits_one = edit_dis_one(word, allow_switches)
    edits_two = set()

    for edit in edits_one:
        edits_two.update(edit_dis_one(edit, allow_switches))

    return edits_two

# Example usage
word = 'bhy'
edits_two = sorted(edit_dis_two(word))

print(f"Number of strings with edit distance of two: {len(edits_two)}")
print(f"First 10 strings: {edits_two[:10]}")
print(f"Last 10 strings: {edits_two[-10:]}")
```

```python
from google.colab import drive
import re
from collections import Counter

# Step 1: Mount Google Drive
drive.mount('/content/drive')

# Step 2: Set the correct file path
file_path = '/content/drive/My Drive/Colab Notebooks/shakespeare.txt'  # Update this path as needed

# Step 0: Function to read and preprocess the shakespeare.txt file
def Your_Read_Function(file_path):
    with open(file_path, 'r') as file:
        text = file.read().lower()
    # Extract words using regex and create a Counter dictionary
    words = re.findall(r'\b\w+\b', text)
    return Counter(words)

# Step 1: Edit distance helper functions
def edit_distance_one(word):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes = [L + R[1:] for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]
    replaces = [L + c + R[1:] for L, R in splits if R for c in letters]
    inserts = [L + c + R for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edit_distance_two(word):
    return set(e2 for e1 in edit_distance_one(word) for e2 in edit_distance_one(e1))

# Step 2: Spelling suggestion function
def fix_edits(word, vocab, n=5):
    suggestions = set()

    # If word is in vocabulary, suggest the word
    if word in vocab:
        suggestions.add(word)

    # If not, get suggestions from edit_distance_one and edit_distance_two
    if not suggestions:
        suggestions_one = edit_distance_one(word)
        suggestions = suggestions_one.intersection(vocab)

    if not suggestions:
        suggestions_two = edit_distance_two(word)
        suggestions = suggestions_two.intersection(vocab)

    # If still no suggestions, return the input word
    if not suggestions:
        return [(word, 0)]

    # Step 3: Calculate probability and sort best words
    best_words = {w: vocab[w] / sum(vocab.values()) for w in suggestions}

    # Return top n suggestions sorted by probability
    return Counter(best_words).most_common(n)

# Example usage
vocab = Your_Read_Function(file_path)

# Testing with an example word
word = "jensi"
suggestions = fix_edits(word, vocab, n=5)

# Print results
print(f"Entered word = {word}")
for i, (suggested_word, prob) in enumerate(suggestions):
    print(f"Word {i}: {suggested_word}, Probability {prob:.6f}")
```
